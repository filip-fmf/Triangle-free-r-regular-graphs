n mora biti liho število, ker je vseh vozlišč n\*r kar mora biti sodo. Ker je r liho \-&gt; n je sodo. S tem zmanjšamo iskanje grafov na polovico \(ne rabimo preverjat vseh n, amapak samo sode\). 



Zaradi teze: \(b\) Concerning \(iii\) with r ≥ 3 \(not necessarily odd\), since there is no requirement on any N\(v\), we must have |G| ≤ r\(r^2 − 1\), moramo za r =3, 5  preverjati samo n &lt;= 24, 120.


Zaradi predolgega delovanja in pomankljivosti RAMA sva se odločila za točko iii\) narisati grafe do n = 16 in samo prešteti koliko ustreznih grafov dobimo za večje n.


```python
#iskanje grafov do n = 16, r = 3, te tudi izrišemo
load("alpha_od.sage")

r = 3
solutions_by_n = {}

# sodi n do 16
for n in range(4, 17, 2):
    print(f"\nIščem n = {n}")

    sols = []
    count = 0

    for i, G in enumerate(graphs.nauty_geng(f"{n} -d3 -D3 -t")):

        # da vidimo da še vedno deluje
        if i % 1000 == 0 and i > 0:
            print(f"  preveril {i} grafov, našel {count}")

        if not G.is_connected():
            continue

        if alpha_od_ilp_correct(G) == r:
            sols.append(G)
            count += 1

    solutions_by_n[n] = sols
    print(f"Konec n = {n}, skupaj našel: {count}")

```

```python
# shranjevanje grafov do n = 16, r = 3
for n in sorted(solutions_by_n):
    graphs = solutions_by_n[n]

    if len(graphs) == 0:
        print(f"Ni grafov za n = {n}")
        continue

    print(f"Rišem n = {n}, {len(graphs)} grafe")

    plots = [
        G.plot(
            vertex_size=30,
            edge_thickness=2,
            vertex_labels=False,
            layout="spring",
            figsize=4
        )
        for G in graphs
    ]

    graphics_array(
        plots,
        ncols=3
    ).save(f"rešitve_n{n}_r3.pdf")
    
    with open("graphs_iii.g6", "w") as f:
        for G in graphs:
            f.write(G.graph6_string() + "\n")
```

```python
# razdeljevanje slik grafov v 3 dele za n = 14 za lepši izgled
import math

n = 14
graphs = solutions_by_n[n]

chunk_size = 8   # koliko grafov na en PDF
num_chunks = math.ceil(len(graphs) / chunk_size)

for i in range(num_chunks):
    chunk = graphs[i*chunk_size:(i+1)*chunk_size]

    plots = [
        G.plot(
            vertex_size=30,
            edge_thickness=2.5,
            vertex_labels=False,
            layout="spring",
            figsize=4.5
        )
        for G in chunk
    ]

    graphics_array(plots, ncols=2).save(f"rešitve_n{n}_r3_del{i+1}.pdf")

    print(f"shranil rešitve_n{n}_del{i+1}.pdf ({len(chunk)} grafov)")
```
